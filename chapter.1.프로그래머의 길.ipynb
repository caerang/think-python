# 형식 언어와 자연 언어

* 언어에는 크게 두 가지 언어가 존재
* 영어, 프랑스어, 중국어와 같이 사람이 일상에서 사용하는 언어 (자연 언어)
* 수학 기호, 화학 기호, 이모티콘과 같이 특정한 용도로 만들어 낸 언어 (형식 언어)
* 프로그래밍 언어도 컴퓨터에 작업을 지시하기 위해 만든 형식 언어
* 형식 언어는 엄격한 구문 규칙을 갖음
* 구문 규칙은 토큰과 구조로 표현
* 토큰은 언어의 기본 요소, 단어, 숫자 등 최소한의 의미를 갖는 단위
